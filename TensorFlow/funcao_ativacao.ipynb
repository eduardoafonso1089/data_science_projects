```python
import numpy as np
```

# Used for linearly separable problems

```python
def stepFunction(sum):
  if sum >= 1:
    return 1
  return 0
```

# Binary classification problems

```python
def sigmoidFunction(sum):
  return 1 / (1 + np.exp(-sum))
```

# Classification values between 1 and -1

```python
def tahnFunction(sum):
  return (np.exp(sum)-np.exp(-sum)) / (np.exp(sum)+np.exp(-sum))
```

# CNN (the most used)

```python
def reluFunction(sum):
  if(sum >=0):
    return sum
  else:
    return 0
```

# Passing value forward

```python
def linearFunction(sum):
  return sum
```

# Classification problem with more than 2 classes

```python
def softmaxFunction(x):
  ex = np.exp(x)
  return ex / ex.sum()
```
