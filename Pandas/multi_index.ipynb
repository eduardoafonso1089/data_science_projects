```python
import pandas as pd
```

```python
arrays = [ [1,1,2,2] , ['red', 'blue', 'red', 'blue']]
```

```python
pd.MultiIndex.from_arrays(arrays, names=['number', 'color'])
```

```python
numbers = [0,1,2]
colors = ['green', 'purple']

pd.MultiIndex.from_product([numbers, colors], names=['number', 'color'])
```
