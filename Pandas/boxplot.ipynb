```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
```

```python
np.random.seed(seed=0)

x = np.random.randn(1000)
y = np.random.randn(100)
z = np.random.randn(10)
```

```python
fig, ax = plt.subplots()

ax.boxplot((x, y, z), labels=('x', 'y', 'z'), vert=False, showmeans=True, meanline=True, patch_artist=True, medianprops={'linewidth':2, 'color':'purple'}, meanprops={'linewidth':2, 'color':'red'})
plt.show()
```
