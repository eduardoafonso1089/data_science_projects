#Função Pivot

```python
dias = pd.date_range('2019-01-01', periods=12)
```

```python
Pessoa = ['George', 'Victor', 'Lucas']
```

```python
np.random.choice(Pessoa)
```

```python
Nome = []
Gasto = []

for i in range(12):
  Nome.append(np.random.choice(Pessoa))
  Gasto.append(np.round(np.random.rand()*100, 2))

Nome
```

```python
Gasto
```

```python
df = pd.DataFrame({'Dia' : dias,'Nome': Nome, 'Gasto': Gasto})
```

```python
df
```

```python
df.pivot(index='Dia', columns='Nome', values='Gasto')
```

#Pivot table

```python
Carros = [7,4,3,2,8]
dias = pd.date_range('2019-01-01', '20190101', periods=5)

vendedor = ['George', 'Vagner', 'Pedro', 'Vagner', 'George']

df = pd.DataFrame({'Vendas': Carros, 'Data': dias, 'Vendedor': vendedor})
```

```python
df
```

```python
pd.pivot(df, values='Vendas', index='Data', columns='Vendedor')
#ERRO!
```

```python
pd.pivot_table(df, values='Vendas', index='Data', columns='Vendedor', aggfunc="sum")
```
